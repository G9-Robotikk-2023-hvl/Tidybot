# matriser eksempel

```python
import numpy as np

x,y,z = 1,2,3

theta=0*np.pi/180
c,s = np.cos(theta), np.sin(theta)

#x rotasjon med translasjon

A = np.array([[0, 0, 1, x],
              [0, c,-s, y],
              [0, s, c, z],
              [0, 0, 0, 1]])

display(A)
```

# noe klot for 책 leke med grafer

```python
import sympy as sp
from sympy.plotting import plot

# definere variabelen
x = sp.symbols("x")

#bare sette paremetrene for a,b,c og d
a,b,c,d=5,0,0,0

#definere funksjonen for f av x
f = a*x**4+b*x**3+c*b**2+c*x**1+d
g = -f

#definere definisjonsomr책det
inner_spann_point=-5
outer_spann_point=5

#plotte( (f, (med x i definisjonsomr책det)) og (g, (med x i definisjonsomr책det))
plot((f,(x,inner_spann_point,outer_spann_point)),((g,(x,inner_spann_point,outer_spann_point))))
```
